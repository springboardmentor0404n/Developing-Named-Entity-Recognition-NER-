```python
import zipfile
import os

# âœ… The zip file name you have
zip_path = "archive (1).zip"

# âœ… Folder to extract everything into
extract_path = "sec_filings_dataset"

# Create the folder if it doesn't exist
os.makedirs(extract_path, exist_ok=True)

# Unzip it
with zipfile.ZipFile(zip_path, "r") as zip_ref:
    zip_ref.extractall(extract_path)

print("âœ… Unzipped successfully!")

# Check what got extracted
print("\nFiles inside 'sec_filings_dataset':")
print(os.listdir(extract_path))
```

```python
import json, os, pandas as pd

# Folder with extracted JSON files
json_folder = os.path.join("sec_filings_dataset", "extracted")

all_reports = []

for file in os.listdir(json_folder):
    path = os.path.join(json_folder, file)
    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)

    all_reports.append({
        "company": data.get("company", ""),
        "filing_date": data.get("filing_date", ""),
        "risk_factors": data.get("item_1A", ""),
        "md&a": data.get("item_7", ""),
        "financial_statements": data.get("item_8", "")
    })

df = pd.DataFrame(all_reports)

# Save to CSV
df.to_csv("financial_report_sections.csv", index=False)
print("âœ… Financial report segmentation complete!")
print("Saved as 'financial_report_sections.csv'")
df.head()
```
